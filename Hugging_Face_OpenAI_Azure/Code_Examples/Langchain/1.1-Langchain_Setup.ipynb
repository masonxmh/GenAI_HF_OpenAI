#### Chains
* `RunnableSequence`
A LangChain runnable that chains steps together in order. The output of one step becomes the input to the next step. It is useful when you want a pipeline like: prompt -> model -> parser.
* `StrOutputParser`
A parser that converts a modelŌĆÖs output into a plain Python string. This is commonly used after an LLM call when you just want the text content.
* `RunnablePassthrough`
A runnable that passes the input through unchanged. It is helpful when you want to keep the original input available while combining it with other
^ `RunnableParallel`
A runnable that executes multiple runnables in parallel on the same input and returns all their results together, usually as a dictionary.



```python
from langchain_core.runnables import RunnableSequence
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableParallel
```

Models
* `ChatOpenAI`
Used for chat-based models. This is the most common choice for modern OpenAI models and works well for conversations, Q&A, agents, and prompt chains built around messages.
* `OpenAI`
Used for older completion-style language model usage. It is less common in newer LangChain projects, since most current workflows use chat models instead.
* `OpenAIEmbeddings`
Used to convert text into vector embeddings. This is useful for semantic search, retrieval, RAG pipelines, similarity matching, and storing vectors in a vector database.



```python
# Models
from langchain_openai import ChatOpenAI, OpenAI, OpenAIEmbeddings
```

#### Documents
`Document` is a LangChain object used to represent a piece of text plus its metadata.

It usually contains:

* `page_content`: the main text
* `metadata`: extra information such as source, file name, page number, URL, or tags

```python
# Documents
from langchain_core.documents import Document
doc = Document(
    page_content="LangChain helps build LLM applications.",
    metadata={"source": "notes.txt", "topic": "llm"}
)
```

#### Prompts
* `PromptTemplate`
Used for standard text prompts with variables.
Good for simple prompt strings like "Summarize this: {text}".
* `ChatPromptTemplate`
Used for chat-style prompts made of messages such as system, human, and assistant roles.
This is the common choice when working with chat models like ChatOpenAI.
* `MessagesPlaceholder`
A placeholder for inserting a list of messages dynamically into a chat prompt.
Useful for chat history, memory, or agent conversation state.
* `HumanMessagePromptTemplate`
Used to define a human/user message template inside a chat prompt.
Helpful when you want explicit control over the user message format.

```python
# Prompts
from langchain_core.prompts import (
    PromptTemplate,
    ChatPromptTemplate,
    MessagesPlaceholder,
    HumanMessagePromptTemplate,
)
```

#### Vector stores

A vector database integration for storing document embeddings and retrieving similar documents.
It is commonly used in RAG workflows and is convenient for local development.
* `FAISS`
A fast similarity search library from Meta, wrapped by LangChain as a vector store.
It is often used for efficient in-memory or local vector search over embeddings.

#### Text splitters
* `CharacterTextSplitter`
Splits text based on character count.
Simple and easy to use, but it may cut text in awkward places.
* `RecursiveCharacterTextSplitter`
A smarter splitter that tries multiple separators like paragraphs, lines, and spaces before falling back to character splitting.
This is one of the most commonly used splitters for RAG.
* `TokenTextSplitter`
Splits text by token count instead of character count.
Better when you want chunk sizes that align more closely with model token limits.
* `SentenceTransformersTokenTextSplitter`
Splits text using tokenization behavior aligned with Sentence Transformers models.
Useful when your embedding workflow depends on sentence-transformer-style token limits.

```python
from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter,
    TokenTextSplitter,
    SentenceTransformersTokenTextSplitter,
)
```
